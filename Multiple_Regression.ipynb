```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("mulitregdata.csv")
df.head()
```

```python
import math
bedroom_median = math.floor(df.bedrooms.median())
print(bedroom_median)
```

```python
df.bedrooms = df.bedrooms.fillna(bedroom_median)
df.head()
```

```python
plt.scatter(df.area, df.price)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Price vs Area")
plt.show()
```

```python
plt.scatter(df.bedrooms, df.price)
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.title("Price vs Bedrooms")
plt.show()
```

```python
plt.scatter(df.age, df.price)
plt.xlabel("Age")
plt.ylabel("Price")
plt.title("Price vs Age")
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred = reg.predict(df[['area', 'bedrooms', 'age']]) # reg.predict is the main prediction function (remember)

plt.scatter(df.price, y_pred)

min_price = min(df.price.min(), y_pred.min())
max_price = max(df.price.max(), y_pred.max())

plt.plot([min_price, max_price], [min_price, max_price])

plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
